```python
import pandas as pd
```

```python
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet')
```

```python
df
```

```python
df.shape
```

```python
df['trip_duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
```

```python
df['trip_duration_min'] = df['trip_duration'].dt.total_seconds() / 60
```

```python
df.head()
```

```python
std_dev = df['trip_duration_min'].std()
print(f"Standard Deviation: {std_dev}")
```

```python
df_filtered = df[(df['trip_duration_min'] >= 1) & (df['trip_duration_min'] <= 60)]
```

```python
df_filtered
```

```python
len(df_filtered)/len(df) * 100
```

```python
df_filtered['PULocationID'] = df_filtered['PULocationID'].astype(str)
df_filtered['DOLocationID'] = df_filtered['DOLocationID'].astype(str)
```

```python
dicts = df_filtered[['PULocationID', 'DOLocationID']].to_dict(orient='records')
```

```python
import numpy as np
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer()
X = dv.fit_transform(dicts)
```

```python
print("Shape of the feature matrix:", X.shape)
```

```python
y = df_filtered['trip_duration_min'].values
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y, y_pred))
print(f"RMSE on training data: {rmse:.2f}")
```
